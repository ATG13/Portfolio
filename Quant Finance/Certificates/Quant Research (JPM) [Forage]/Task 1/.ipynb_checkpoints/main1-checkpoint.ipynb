## Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
import datetime as dt
```

## Preprocessing the data

```python
data = pd.read_csv("Nat_Gas.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data['Dates'] = pd.to_datetime(data['Dates'])
```

```python
# Figure
plt.figure(figsize=(12, 6))
plt.plot(data["Dates"], data["Prices"],  marker='o', c='g')
plt.title('Price of Natural Gas over time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
```

Notes:
- We can see that there is a general upwards trend in the data.
- There is also an aspect of seasonality, we see the prices tend to be higher during the first 6 months of the year and then falls towards the end of the year.

We need to predict the time series for the next 1 year. Knowing that there is seasonality and a general trend in the data.

Also am advised to look at the factors that might affect the pricing. (But within the dataset only)

## Model attempt

**Objective:** input(date) = output(price value)

Linear regression on past data to predict future one year data.

```python
# Getting month and year data
data['Year'] = data['Dates'].dt.year
data['Month'] = data['Dates'].dt.month
```

```python
data.head()
```

```python
# Np array for each month
prices = []
for i in np.arange(12):
    X = np.array(data[data['Month'] == i+1]['Year']).reshape(-1,1)
    y = np.array(data[data['Month'] == i+1]['Prices'])

    reg = LinearRegression().fit(X,y)
    price = round(float(reg.predict([[2025]])),2)
    prices.append(price)
```

```python
prices = np.array(prices)
prices
```

```python
data.tail()
```

```python
start_date = '2024-10-01'
end_date = '2025-09-30'


date_range = pd.date_range(start=start_date, end=end_date, freq='D')

last_days_of_month = date_range.to_period('M').to_timestamp('M')

last_days = np.array(np.unique(last_days_of_month.values), dtype='datetime64[D]')

last_days
```

```python
data1 = {
    'Dates': last_days,
    'Prices': prices
}
data1 = pd.DataFrame(data1)
data1
```

```python
# Figure
plt.figure(figsize=(12, 6))
plt.plot(data["Dates"], data["Prices"],  marker='o', c='g')
plt.plot(data1["Dates"], data1["Prices"],  marker='x', c='b')
plt.title('Prediction of natural gas price over time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
```

Simple model based on linear regression of the previous year data
